```python
titanic = df[['survived','pclass','sex','age','sibsp','parch','fare','embark_town','alone']].copy()
```

```python
titanic.isnull().sum().plot(kind='pie',figsize=(8,3), autopct='%f%%')
```

fix missing value
 numeric: mean
 categorical: mode
 

```python
avg_age = round(titanic['age'].mean())
print(f'Average age: {avg_age: .1f}')
```

```python
titanic['age'].fillna(avg_age,inplace=True)
```

```python
# most frequent value in embark_town column
titanic['embark_town'].value_counts().idxmax()
```

```python
most_freq_embark_town = titanic['embark_town'].value_count().idxmax()
titanic ['embark_town'].fillna(most_freq_town, inplace=True)
```
