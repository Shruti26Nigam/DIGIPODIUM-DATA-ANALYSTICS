```python
class Triangle:
    def __init__(self, b=0, h=0):
        self.base = b
        self.height = h

    def area(self):
        return self.base * self.height / 2
    
    def __repr__(self):
        return f'Triangle({self.area()})'
```

```python
data = [Triangle(12,3), Triangle(34,23), Triangle(8,15), Triangle(6,25)]
data
```
