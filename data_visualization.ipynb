```python
sns.histplot(iris, x='sepal_length', bins=15,kde=True) #density
```

```python
# to create multiple graph data in single graph
sns.set_palette('rainbow')
fig, ax =plt.subplots(figsize=(20,5), ncols = 4)
sns.histplot(iris, x = 'sepal_width', bins = 15, kde = True, ax=ax[0], color='blue')
sns.histplot(iris, x = 'petal_width', bins = 15, kde = True, ax=ax[1])
sns.histplot(iris, x = 'sepal_length', bins = 15, kde = True, ax=ax[2])
sns.histplot(iris, x = 'petal_length', bins = 15, kde = True, ax=ax[3],color='orange')
```

```python
sns.set_palette('rainbow')
fig, ax =plt.subplots(figsize=(20,8), ncols = 2,nrows = 2)
sns.histplot(iris, x = 'sepal_width', bins = 15, kde = True, ax=ax[0,0], color='blue')
sns.histplot(iris, x = 'petal_width', bins = 15, kde = True, ax=ax[0,1])
sns.histplot(iris, x = 'sepal_length', bins = 15, kde = True, ax=ax[1,0])
sns.histplot(iris, x = 'petal_length', bins = 15, kde = True, ax=ax[1,1],color='orange')
```
